```python
# Helper Functions for Poisoned Data

import random

def should_apply_poisoned_data(agent_name: str, config: dict) -> bool:
    """
    Check if agent should receive poisoned data based on config.
    
    Args:
        agent_name: Name of the agent
        config: Poisoned data configuration dictionary
    
    Returns:
        bool: True if poisoned data should be applied, False otherwise
    """
    if not config.get("enabled", False):
        return False
    
    # Check if agent is in poisoned_agents config
    poisoned_agents = config.get("poisoned_agents", {})
    if agent_name not in poisoned_agents:
        return False
    
    agent_config = poisoned_agents[agent_name]
    
    # Check probability if configured
    probability = agent_config.get("probability", 1.0)
    if probability < 1.0:
        return random.random() < probability
    
    return True


def get_poisoned_token_counts(agent_name: str, config: dict, 
                              base_prompt: int, base_completion: int) -> tuple:
    """
    Calculate high token counts for poisoned agents.
    
    Args:
        agent_name: Name of the agent
        config: Poisoned data configuration dictionary
        base_prompt: Base prompt token count
        base_completion: Base completion token count
    
    Returns:
        tuple: (prompt_tokens, completion_tokens, total_tokens)
    """
    poisoned_agents = config.get("poisoned_agents", {})
    agent_config = poisoned_agents.get(agent_name, {})
    default_high = config.get("default_high_tokens", {})
    
    # Use token_multiplier if specified
    if "token_multiplier" in agent_config:
        multiplier = agent_config["token_multiplier"]
        prompt_tokens = int(base_prompt * multiplier)
        completion_tokens = int(base_completion * multiplier)
    # Use min_tokens if specified
    elif "min_tokens" in agent_config:
        min_tokens = agent_config["min_tokens"]
        prompt_tokens = max(min_tokens, base_prompt)
        completion_tokens = max(min_tokens // 2, base_completion)
    # Use default_high_tokens
    else:
        prompt_tokens = default_high.get("prompt", base_prompt * 10)
        completion_tokens = default_high.get("completion", base_completion * 10)
    
    total_tokens = prompt_tokens + completion_tokens
    
    return (prompt_tokens, completion_tokens, total_tokens)
```

```python
# Timer utility function
from contextlib import contextmanager
import time

@contextmanager
def timer():
    """Context manager to time execution"""
    start_time = time.time()
    yield
    end_time = time.time()
    execution_time = end_time - start_time
    print(f"Execution time: {execution_time:.2f} seconds")
```

```python
# Helper function for creating LLM calls with tools
def create_llm_call_with_tools(
    tracer,
    llm_span_name: str,
    system_message: str,
    user_message: str,
    template: str,
    template_version: str,
    scenario: str,
    tools: list,
    is_poisoned: bool,
    poisoned_config: dict,
    agent_name: str,
    session_id: str = None,
    doc_content: str = None
):
    """
    Helper function to create an LLM call with potential tool usage.
    
    Returns:
        tuple: (total_tokens, prompt_tokens, completion_tokens, final_response)
    """
    from opentelemetry.trace.status import Status, StatusCode
    import json
    import time
    
    with tracer.start_as_current_span(llm_span_name) as llm_span:
        if session_id:
            llm_span.set_attribute("session.id", session_id)
        
        # Calculate token counts
        if is_poisoned:
            prompt_tokens, completion_tokens, total_tokens = get_poisoned_token_counts(
                agent_name, poisoned_config, base_prompt=1250, base_completion=450
            )
        else:
            prompt_tokens = random.randint(1000, 2000)
            completion_tokens = random.randint(300, 800)
            total_tokens = prompt_tokens + completion_tokens
        
        # Calculate latency
        if is_poisoned:
            llm_latency_ms = random.uniform(4000, 8000)
        else:
            llm_latency_ms = random.uniform(1500, 4000)
        
        # Set LLM basic attributes
        llm_span.set_attribute("openinference.span.kind", "LLM")
        llm_span.set_attribute("llm.model_name", "gemini-1.5-pro")
        llm_span.set_attribute("llm.provider", "google")
        llm_span.set_attribute("llm.system", "google")
        
        # Set prompt template attributes
        llm_span.set_attribute("llm.prompt_template.template", template)
        llm_span.set_attribute("llm.prompt_template.variables", json.dumps({"user_query": user_message}))
        llm_span.set_attribute("llm.prompt_template.version", template_version)
        
        # Set tools if available
        tool_calls = []
        if tools:
            tools_json = json.dumps([{
                "type": "function",
                "function": {
                    "name": tool["name"],
                    "description": tool["description"],
                    "parameters": tool["parameters"]
                }
            } for tool in tools])
            llm_span.set_attribute("llm.tools", tools_json)
            
            # Generate tool calls (higher probability for more tool usage)
            for tool_config in tools:
                if random.random() < 0.85:  # 85% chance each tool is called
                    tool_calls.append(generate_tool_call(tool_config, scenario))
        
        # Set input messages
        llm_span.set_attribute("llm.input_messages.0.message.role", "system")
        llm_span.set_attribute("llm.input_messages.0.message.content", system_message)
        llm_span.set_attribute("llm.input_messages.1.message.role", "user")
        llm_span.set_attribute("llm.input_messages.1.message.content", user_message)
        
        # Generate response (simplified - in real scenario this would be from LLM)
        response = f"Agent response for: {user_message[:50]}..."
        
        # Add tool calls to output if present
        if tool_calls:
            tool_calls_json = json.dumps(tool_calls)
            llm_span.set_attribute("llm.output_messages.0.message.tool_calls", tool_calls_json)
            
            for idx, tool_call in enumerate(tool_calls):
                llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.id", tool_call["id"])
                llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.function.name", tool_call["function"]["name"])
                llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.function.arguments", tool_call["function"]["arguments"])
            
            # Create TOOL child spans for each tool call
            for tool_call in tool_calls:
                tool_name = tool_call["function"]["name"]
                tool_args = json.loads(tool_call["function"]["arguments"])
                
                with tracer.start_as_current_span(f"tool_{tool_name}") as tool_span:
                    tool_span.set_attribute("openinference.span.kind", "TOOL")
                    tool_span.set_attribute("tool.name", tool_name)
                    tool_span.set_attribute("tool.parameters", tool_call["function"]["arguments"])
                    tool_span.set_attribute("input.value", tool_call["function"]["arguments"])
                    tool_span.set_attribute("input.mime_type", "application/json")
                    
                    tool_response = generate_tool_response(tool_name, tool_args)
                    tool_span.set_attribute("output.value", json.dumps(tool_response))
                    tool_span.set_attribute("output.mime_type", "application/json")
                    
                    tool_latency_ms = random.uniform(50, 500)
                    tool_span.set_attribute("latency_ms", round(tool_latency_ms, 2))
                    time.sleep(0.001)
                    tool_span.set_status(Status(StatusCode.OK))
                    
                    # Update response with tool results
                    response += f" [Tool {tool_name} executed successfully]"
        else:
            # Set output message without tool calls
            llm_span.set_attribute("llm.output_messages.0.message.role", "assistant")
            llm_span.set_attribute("llm.output_messages.0.message.content", response)
        
        # Set token counts
        llm_span.set_attribute("llm.token_count.prompt", prompt_tokens)
        llm_span.set_attribute("llm.token_count.completion", completion_tokens)
        llm_span.set_attribute("llm.token_count.total", total_tokens)
        
        # Set invocation parameters
        invocation_params = {
            "temperature": 0.7,
            "max_tokens": 2000,
            "top_p": 0.9,
            "frequency_penalty": 0.0,
            "presence_penalty": 0.0
        }
        llm_span.set_attribute("llm.invocation_parameters", json.dumps(invocation_params))
        
        # Set input/output values
        llm_span.set_attribute("input.value", user_message)
        llm_span.set_attribute("input.mime_type", "text/plain")
        llm_span.set_attribute("output.value", response)
        llm_span.set_attribute("output.mime_type", "text/plain")
        
        # Set retrieval documents if applicable
        if doc_content:
            DOCUMENT_SIZE = 10000
            document_content = doc_content * (DOCUMENT_SIZE // len(doc_content) + 1)
            document_content = document_content[:DOCUMENT_SIZE]
            llm_span.set_attribute("retrieval.documents.0.document.content", document_content)
            llm_span.set_attribute("retrieval.documents.0.document.id", f"doc_{scenario or 'default'}")
            llm_span.set_attribute("retrieval.documents.0.document.score", 0.75)
        
        llm_span.set_status(Status(StatusCode.OK))
        llm_span.set_attribute("latency_ms", round(llm_latency_ms, 2))
        time.sleep(0.01)
        
        return total_tokens, prompt_tokens, completion_tokens, response
```

```python
# Complex Multi-Agent Trace Function
# Creates traces with multiple agents, multiple LLM calls, and extensive tool usage
# Structure: CHAIN -> routing_agent -> specialist_agent -> (optional) confirmation_agent

def create_single_trace_complex(
    tracer,
    trace_id: str = None,
    agent_name: str = "customer_service_agent",
    user_query: str = None,
    call_center_scenario: str = None,
    prompt_template_config: dict = None,
    poisoned_config: dict = None,
    session_id: str = None
):
    """
    Create a complex trace with multiple agents, LLM calls, and tool usage.
    
    Structure: CHAIN -> routing_agent -> specialist_agent -> (optional) confirmation_agent
    Each agent can have multiple LLM calls with tool usage.
    """
    from opentelemetry.trace.status import Status, StatusCode
    import json
    import time
    
    # Get call center data based on scenario
    query, response, doc_content = get_call_center_data(call_center_scenario)
    
    if user_query:
        query = user_query
    
    # Determine workflow complexity (2-4 agents)
    num_agents = random.choice([2, 3, 3, 4])  # Weighted toward 3 agents
    
    # Determine which agents to use based on scenario
    if call_center_scenario in ["account_cancellation", "account_retention"]:
        specialist_agent = "retention_agent"
    elif call_center_scenario == "sales":
        specialist_agent = "sales_agent"
    else:
        specialist_agent = "customer_service_agent"
    
    # Check if any agent should be poisoned
    is_poisoned = should_apply_poisoned_data(specialist_agent, poisoned_config or {})
    
    chain_name = "TelcoCo Call Center"
    total_prompt_tokens = 0
    total_completion_tokens = 0
    final_response = response
    
    with tracer.start_as_current_span(chain_name) as chain_span:
        chain_span.set_attribute("openinference.span.kind", "CHAIN")
        chain_span.set_attribute("input.value", query)
        chain_span.set_attribute("input.mime_type", "text/plain")
        
        if session_id:
            chain_span.set_attribute("session.id", session_id)
        
        time.sleep(0.01)
        
        # AGENT 1: Routing/Initial Agent
        routing_template = prompt_template_config.get("opening_greeting", {}) if prompt_template_config else {}
        routing_system = routing_template.get("system_message", "You are a TelcoCo call center routing agent.")
        routing_template_str = routing_template.get("template", "{user_query}")
        routing_version = routing_template.get("version", "v1.0")
        
        with tracer.start_as_current_span("routing_agent") as routing_span:
            routing_span.set_attribute("openinference.span.kind", "AGENT")
            routing_span.set_attribute("agent.name", "routing_agent")
            routing_span.set_attribute("input.value", query)
            routing_span.set_attribute("description", f"Routing agent processing customer request: {query[:100]}")
            
            if session_id:
                routing_span.set_attribute("session.id", session_id)
            
            # First LLM call: Initial understanding
            routing_tools = get_random_tools(num_tools=random.randint(1, 2), scenario="opening_greeting")
            tokens1, prompt1, comp1, resp1 = create_llm_call_with_tools(
                tracer, "llm_call_initial", routing_system,
                routing_template_str.format(user_query=query), routing_template_str, routing_version,
                "opening_greeting", routing_tools, False, poisoned_config, "routing_agent", session_id
            )
            total_prompt_tokens += prompt1
            total_completion_tokens += comp1
            
            # Second LLM call: Routing decision
            routing_decision_msg = f"Based on: {query[:100]}, determine the best specialist agent to handle this."
            tokens2, prompt2, comp2, resp2 = create_llm_call_with_tools(
                tracer, "llm_call_routing", routing_system, routing_decision_msg,
                routing_template_str, routing_version, "opening_greeting", [], False,
                poisoned_config, "routing_agent", session_id
            )
            total_prompt_tokens += prompt2
            total_completion_tokens += comp2
            
            routing_span.set_attribute("latency_ms", round(random.uniform(100, 300), 2))
            time.sleep(0.01)
        
        # AGENT 2: Specialist Agent
        specialist_template = prompt_template_config.get(call_center_scenario or "customer_service", {}) if prompt_template_config else {}
        specialist_system = specialist_template.get("system_message", "You are a TelcoCo specialist agent.")
        specialist_template_str = specialist_template.get("template", "{user_query}")
        specialist_version = specialist_template.get("version", "v1.0")
        
        with tracer.start_as_current_span(specialist_agent) as specialist_span:
            specialist_span.set_attribute("openinference.span.kind", "AGENT")
            specialist_span.set_attribute("agent.name", specialist_agent)
            specialist_span.set_attribute("input.value", query)
            specialist_span.set_attribute("description", f"{specialist_agent.replace('_', ' ').title()} handling customer request: {query[:100]}")
            
            if session_id:
                specialist_span.set_attribute("session.id", session_id)
            
            # First LLM call: Initial specialist response with tools
            specialist_tools = get_random_tools(num_tools=random.randint(2, 4), scenario=call_center_scenario)
            tokens3, prompt3, comp3, resp3 = create_llm_call_with_tools(
                tracer, "llm_call_specialist_initial", specialist_system,
                specialist_template_str.format(user_query=query), specialist_template_str, specialist_version,
                call_center_scenario, specialist_tools, is_poisoned, poisoned_config,
                specialist_agent, session_id, doc_content
            )
            total_prompt_tokens += prompt3
            total_completion_tokens += comp3
            final_response = resp3
            
            # Second LLM call: Follow-up with tool results
            if specialist_tools and random.random() < 0.8:  # 80% chance of follow-up
                followup_msg = f"Previous response: {resp3[:200]}. Provide additional details or next steps."
                followup_tools = get_random_tools(num_tools=random.randint(1, 2), scenario=call_center_scenario)
                tokens4, prompt4, comp4, resp4 = create_llm_call_with_tools(
                    tracer, "llm_call_specialist_followup", specialist_system, followup_msg,
                    specialist_template_str, specialist_version, call_center_scenario,
                    followup_tools, is_poisoned, poisoned_config, specialist_agent, session_id
                )
                total_prompt_tokens += prompt4
                total_completion_tokens += comp4
                final_response = resp4
            
            specialist_span.set_attribute("latency_ms", round(random.uniform(200, 500), 2))
            time.sleep(0.01)
        
        # AGENT 3: Optional Follow-up/Confirmation Agent (30% chance)
        if num_agents >= 3 and random.random() < 0.3:
            confirmation_template = prompt_template_config.get("customer_service", {}) if prompt_template_config else {}
            conf_system = confirmation_template.get("system_message", "You are a TelcoCo confirmation agent.")
            conf_template_str = confirmation_template.get("template", "{user_query}")
            conf_version = confirmation_template.get("version", "v1.0")
            
            with tracer.start_as_current_span("confirmation_agent") as conf_span:
                conf_span.set_attribute("openinference.span.kind", "AGENT")
                conf_span.set_attribute("agent.name", "confirmation_agent")
                conf_span.set_attribute("input.value", query)
                conf_span.set_attribute("description", f"Confirmation agent verifying resolution for: {query[:100]}")
                
                if session_id:
                    conf_span.set_attribute("session.id", session_id)
                
                conf_msg = f"Confirm the resolution: {final_response[:200]}"
                conf_tools = get_random_tools(num_tools=random.randint(0, 1), scenario=call_center_scenario)
                tokens5, prompt5, comp5, resp5 = create_llm_call_with_tools(
                    tracer, "llm_call_confirmation", conf_system, conf_msg,
                    conf_template_str, conf_version, call_center_scenario, conf_tools,
                    False, poisoned_config, "confirmation_agent", session_id
                )
                total_prompt_tokens += prompt5
                total_completion_tokens += comp5
                final_response = resp5
                
                conf_span.set_attribute("latency_ms", round(random.uniform(50, 150), 2))
                time.sleep(0.01)
        
        # Set CHAIN output
        chain_span.set_attribute("output.value", final_response)
        chain_span.set_attribute("output.mime_type", "text/plain")
        chain_span.set_status(Status(StatusCode.OK))
        chain_span.set_attribute("latency_ms", round(random.uniform(2000, 6000), 2))
    
    return {
        "agent_name": specialist_agent,
        "is_poisoned": is_poisoned,
        "token_counts": {
            "prompt": total_prompt_tokens,
            "completion": total_completion_tokens,
            "total": total_prompt_tokens + total_completion_tokens
        }
    }

# Replace the old function with the new complex one
create_single_trace = create_single_trace_complex
```

```python
# Batch generation using the single trace function
# This cell generates multiple traces using the reusable create_single_trace() function

from opentelemetry import trace

tracer = trace.get_tracer(__name__)

# Number of traces to create - adjust as needed
NUM_SPANS = 1000

print(f"Generating {NUM_SPANS:,} complex traces (CHAIN -> Multiple AGENTS -> Multiple LLM calls -> Multiple TOOL calls) with BatchSpanProcessor...")
print("Note: Each trace contains 2-4 agents, each with 1-2 LLM calls, and extensive tool usage.\n")

with timer():
    poisoned_count = 0
    
    # Session management: Create small groups of traces with shared session IDs
    # Each session contains 2-5 traces
    SESSION_SIZE_MIN = 2
    SESSION_SIZE_MAX = 5
    current_session_id = None
    traces_in_current_session = 0
    session_size = 0
    
    # Create traces in batches to show progress and manage memory
    BATCH_SIZE = 10  # Process in batches of 10 for progress updates
    
    for batch_start in range(0, NUM_SPANS, BATCH_SIZE):
        batch_end = min(batch_start + BATCH_SIZE, NUM_SPANS)
        
        for i in range(batch_start, batch_end):
            # Start a new session if needed
            if current_session_id is None or traces_in_current_session >= session_size:
                # Generate new session ID
                current_session_id = f"session_{random.randint(100000, 999999)}"
                session_size = random.randint(SESSION_SIZE_MIN, SESSION_SIZE_MAX)
                traces_in_current_session = 0
            
            # Split traffic across different TelcoCo call center scenarios
            scenario = random.choice(["opening_greeting", "number_porting", "account_cancellation", 
                                     "account_retention", "customer_service", "sales"])
            
            # Determine agent name based on scenario
            if scenario in ["account_cancellation", "account_retention"]:
                agent_name = "retention_agent"
            elif scenario == "sales":
                agent_name = "sales_agent"
            else:
                agent_name = "customer_service_agent"
            
            # Create trace using reusable function with session ID
            result = create_single_trace(
                tracer=tracer,
                agent_name=agent_name,
                call_center_scenario=scenario,
                prompt_template_config=PROMPT_TEMPLATE_CONFIG,
                poisoned_config=POISONED_DATA_CONFIG,
                session_id=current_session_id
            )
            
            traces_in_current_session += 1
            
            if result["is_poisoned"]:
                poisoned_count += 1
        
        # Progress update
        if (batch_end % 100 == 0) or (batch_end == NUM_SPANS):
            print(f"  Created {batch_end:,} / {NUM_SPANS:,} traces ({poisoned_count} poisoned)...")
    
    print(f"\nCompleted creating {NUM_SPANS:,} traces. Flushing remaining spans...")
    
    # Force flush to ensure all spans are exported
    # Calculate timeout based on number of spans
    base_timeout = 30000  # 30 seconds base
    timeout_per_10k_spans = 10000  # +10 seconds per 10k spans
    flush_timeout = base_timeout + (NUM_SPANS // 10000) * timeout_per_10k_spans
    flush_timeout = min(flush_timeout, 300000)  # Cap at 5 minutes max
    
    print(f"Flushing with timeout: {flush_timeout/1000:.0f} seconds")
    flush_success = trace.get_tracer_provider().force_flush(timeout_millis=flush_timeout)
    
    if flush_success:
        print(f"✅ Flush successful: All traces were exported within {flush_timeout/1000:.0f} seconds")
        print(f"   Total: {NUM_SPANS:,} traces, {poisoned_count} with poisoned data")
        print("   Note: This confirms traces were sent to Arize. Check the Arize dashboard to verify receipt.")
    else:
        print(f"⚠️  Flush timeout: Some traces may not have been exported within {flush_timeout/1000:.0f} seconds")
        print("   Consider increasing the timeout or reducing batch size.")
```
