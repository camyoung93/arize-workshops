```python
# Helper Functions for Poisoned Data

import random

def should_apply_poisoned_data(agent_name: str, config: dict) -> bool:
    """
    Check if agent should receive poisoned data based on config.
    
    Args:
        agent_name: Name of the agent
        config: Poisoned data configuration dictionary
    
    Returns:
        bool: True if poisoned data should be applied, False otherwise
    """
    if not config.get("enabled", False):
        return False
    
    # Check if agent is in poisoned_agents config
    poisoned_agents = config.get("poisoned_agents", {})
    if agent_name not in poisoned_agents:
        return False
    
    agent_config = poisoned_agents[agent_name]
    
    # Check probability if configured
    probability = agent_config.get("probability", 1.0)
    if probability < 1.0:
        return random.random() < probability
    
    return True


def get_poisoned_token_counts(agent_name: str, config: dict, 
                              base_prompt: int, base_completion: int) -> tuple:
    """
    Calculate high token counts for poisoned agents.
    
    Args:
        agent_name: Name of the agent
        config: Poisoned data configuration dictionary
        base_prompt: Base prompt token count
        base_completion: Base completion token count
    
    Returns:
        tuple: (prompt_tokens, completion_tokens, total_tokens)
    """
    poisoned_agents = config.get("poisoned_agents", {})
    agent_config = poisoned_agents.get(agent_name, {})
    default_high = config.get("default_high_tokens", {})
    
    # Use token_multiplier if specified
    if "token_multiplier" in agent_config:
        multiplier = agent_config["token_multiplier"]
        prompt_tokens = int(base_prompt * multiplier)
        completion_tokens = int(base_completion * multiplier)
    # Use min_tokens if specified
    elif "min_tokens" in agent_config:
        min_tokens = agent_config["min_tokens"]
        prompt_tokens = max(min_tokens, base_prompt)
        completion_tokens = max(min_tokens // 2, base_completion)
    # Use default_high_tokens
    else:
        prompt_tokens = default_high.get("prompt", base_prompt * 10)
        completion_tokens = default_high.get("completion", base_completion * 10)
    
    total_tokens = prompt_tokens + completion_tokens
    
    return (prompt_tokens, completion_tokens, total_tokens)
```

```python
# Single Trace Function
# Creates a complete trace with CHAIN -> AGENT -> LLM hierarchy

from opentelemetry import trace
from opentelemetry.trace.status import Status, StatusCode
import json
import time

def create_single_trace(
    tracer,
    trace_id: str = None,
    agent_name: str = "financial_analysis_agent",
    user_query: str = None,
    financial_scenario: str = None,
    prompt_template_config: dict = None,
    poisoned_config: dict = None,
    session_id: str = None
):
    """
    Create a single trace with CHAIN -> AGENT -> LLM hierarchy.
    
    Args:
        tracer: OpenTelemetry tracer instance
        trace_id: Optional trace ID (not used, kept for compatibility)
        agent_name: Name of the agent
        user_query: Optional user query (if None, will be generated from scenario)
        financial_scenario: One of "stock_analysis", "transaction_processing", "risk_assessment", "portfolio_analysis"
        prompt_template_config: Prompt template configuration dictionary
        poisoned_config: Poisoned data configuration dictionary
        session_id: Optional session ID to group traces together
    
    Returns:
        dict with trace metadata (e.g., was_poisoned, token_counts, etc.)
    """
    # Get financial data based on scenario
    query, response, doc_content = get_financial_data(financial_scenario)
    
    # Override with user_query if provided
    if user_query:
        query = user_query
    
    # Get prompt template for scenario
    template_config = {}
    if prompt_template_config:
        template_config = prompt_template_config.get(financial_scenario or "stock_analysis", {})
    
    system_message = template_config.get("system_message", "You are a helpful financial assistant.")
    template = template_config.get("template", "{user_query}")
    template_version = template_config.get("version", "v1.0")
    
    # Format user message from template
    user_message = template.format(user_query=query)
    
    # Get tools for this LLM call
    tools = get_random_tools(scenario=financial_scenario)
    tool_calls = []
    if tools:
        # Generate tool calls (some tools may be called)
        for tool_config in tools:
            if random.random() < 0.7:  # 70% chance each tool is actually called
                tool_calls.append(generate_tool_call(tool_config, financial_scenario))
    
    # Check if poisoned
    is_poisoned = should_apply_poisoned_data(agent_name, poisoned_config or {})
    
    # Use generic chain name for all traces
    chain_name = "Customer Chatbot"
    
    # Determine agent span name based on agent type
    if "customer_service" in agent_name.lower():
        agent_span_name = "customer_service_agent"
    else:
        agent_span_name = "financial_agent"
    
    # Calculate realistic latencies (in milliseconds)
    # LLM latency: 500-3000ms for normal, higher for poisoned
    if is_poisoned:
        llm_latency_ms = random.uniform(4000, 8000)  # 4-8 seconds for poisoned
    else:
        llm_latency_ms = random.uniform(1500, 4000)  # 1.5-4 seconds for normal
    
    # Agent latency: 50-200ms (orchestration overhead)
    agent_latency_ms = random.uniform(50, 200)
    
    # Chain latency: sum of agent + LLM + small overhead (20-100ms)
    chain_overhead_ms = random.uniform(20, 100)
    chain_latency_ms = agent_latency_ms + llm_latency_ms + chain_overhead_ms
    
    # Create nested spans with realistic timing
    # Use minimal delays to simulate latency without significantly slowing batch generation
    chain_start = time.time()
    with tracer.start_as_current_span(chain_name) as chain_span:
        # Set CHAIN attributes
        chain_span.set_attribute("openinference.span.kind", "CHAIN")
        chain_span.set_attribute("input.value", query)
        chain_span.set_attribute("input.mime_type", "text/plain")
        
        # Set session ID if provided (unifies traces into groups)
        if session_id:
            chain_span.set_attribute("session.id", session_id)
        
        # Minimal delay to simulate agent orchestration (0.01s = 10ms)
        time.sleep(0.01)
        
        agent_start = time.time()
        with tracer.start_as_current_span(agent_span_name) as agent_span:
            # Set AGENT attributes
            agent_span.set_attribute("openinference.span.kind", "AGENT")
            agent_span.set_attribute("agent.name", agent_name)
            
            # Set session ID if provided
            if session_id:
                agent_span.set_attribute("session.id", session_id)
            
            # Minimal delay to simulate LLM call (0.01s = 10ms)
            time.sleep(0.01)
            
            llm_start = time.time()
            with tracer.start_as_current_span("llm_call") as llm_span:
                # Set session ID if provided
                if session_id:
                    llm_span.set_attribute("session.id", session_id)
                
                # Calculate token counts
                if is_poisoned:
                    prompt_tokens, completion_tokens, total_tokens = get_poisoned_token_counts(
                        agent_name, poisoned_config, base_prompt=1250, base_completion=450
                    )
                else:
                    prompt_tokens = 1250
                    completion_tokens = 450
                    total_tokens = 1700
                
                # Set LLM basic attributes
                llm_span.set_attribute("openinference.span.kind", "LLM")
                llm_span.set_attribute("llm.model_name", "gpt-4.1-2025-04-14")
                llm_span.set_attribute("llm.provider", "openai")
                llm_span.set_attribute("llm.system", "openai")
                
                # Set prompt template attributes
                llm_span.set_attribute("llm.prompt_template.template", template)
                llm_span.set_attribute("llm.prompt_template.variables", json.dumps({"user_query": query}))
                llm_span.set_attribute("llm.prompt_template.version", template_version)
                
                # Set tools if available
                if tools:
                    tools_json = json.dumps([{
                        "type": "function",
                        "function": {
                            "name": tool["name"],
                            "description": tool["description"],
                            "parameters": tool["parameters"]
                        }
                    } for tool in tools])
                    llm_span.set_attribute("llm.tools", tools_json)
                
                # Set input messages (system and user)
                llm_span.set_attribute("llm.input_messages.0.message.role", "system")
                llm_span.set_attribute("llm.input_messages.0.message.content", system_message)
                llm_span.set_attribute("llm.input_messages.1.message.role", "user")
                llm_span.set_attribute("llm.input_messages.1.message.content", user_message)
                
                # Set output messages (assistant response)
                llm_span.set_attribute("llm.output_messages.0.message.role", "assistant")
                llm_span.set_attribute("llm.output_messages.0.message.content", response)
                
                # Add tool calls to output message if present
                if tool_calls:
                    tool_calls_json = json.dumps(tool_calls)
                    llm_span.set_attribute("llm.output_messages.0.message.tool_calls", tool_calls_json)
                    
                    # Also set individual tool call attributes
                    for idx, tool_call in enumerate(tool_calls):
                        llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.id", tool_call["id"])
                        llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.function.name", tool_call["function"]["name"])
                        llm_span.set_attribute(f"llm.output_messages.0.message.tool_calls.{idx}.tool_call.function.arguments", tool_call["function"]["arguments"])
                    
                    # Set function call attribute (for backward compatibility)
                    if len(tool_calls) > 0:
                        first_call = tool_calls[0]
                        llm_span.set_attribute("llm.function_call", json.dumps({
                            "name": first_call["function"]["name"],
                            "arguments": first_call["function"]["arguments"]
                        }))
                    
                    # Create TOOL child spans for each tool call
                    for tool_call in tool_calls:
                        tool_name = tool_call["function"]["name"]
                        tool_args = json.loads(tool_call["function"]["arguments"])
                        
                        with tracer.start_as_current_span(f"tool_{tool_name}") as tool_span:
                            # Set TOOL span attributes
                            tool_span.set_attribute("openinference.span.kind", "TOOL")
                            tool_span.set_attribute("tool.name", tool_name)
                            tool_span.set_attribute("tool.parameters", tool_call["function"]["arguments"])
                            
                            # Set input/output values for tool
                            tool_span.set_attribute("input.value", tool_call["function"]["arguments"])
                            tool_span.set_attribute("input.mime_type", "application/json")
                            
                            # Generate realistic tool response based on tool type
                            tool_response = generate_tool_response(tool_name, tool_args)
                            tool_span.set_attribute("output.value", json.dumps(tool_response))
                            tool_span.set_attribute("output.mime_type", "application/json")
                            
                            # Set tool latency (tools typically take 50-500ms)
                            tool_latency_ms = random.uniform(50, 500)
                            tool_span.set_attribute("latency_ms", round(tool_latency_ms, 2))
                            
                            # Minimal delay to simulate tool execution
                            time.sleep(0.001)  # 1ms delay
                            
                            tool_span.set_status(Status(StatusCode.OK))
                
                # Set token counts
                llm_span.set_attribute("llm.token_count.prompt", prompt_tokens)
                llm_span.set_attribute("llm.token_count.completion", completion_tokens)
                llm_span.set_attribute("llm.token_count.total", total_tokens)
                
                # Set invocation parameters
                invocation_params = {
                    "temperature": 0.7,
                    "max_tokens": 2000,
                    "top_p": 0.9,
                    "frequency_penalty": 0.0,
                    "presence_penalty": 0.0
                }
                llm_span.set_attribute("llm.invocation_parameters", json.dumps(invocation_params))
                
                # Set input/output values (for backward compatibility)
                llm_span.set_attribute("input.value", user_message)
                llm_span.set_attribute("input.mime_type", "text/plain")
                llm_span.set_attribute("output.value", response)
                llm_span.set_attribute("output.mime_type", "text/plain")
                
                # Set retrieval documents if applicable
                if doc_content:
                    # Generate document content - REDUCED SIZE to avoid timeout issues
                    DOCUMENT_SIZE = 10000  # 10k characters per document
                    document_content = doc_content * (DOCUMENT_SIZE // len(doc_content) + 1)
                    document_content = document_content[:DOCUMENT_SIZE]
                    
                    llm_span.set_attribute("retrieval.documents.0.document.content", document_content)
                    llm_span.set_attribute("retrieval.documents.0.document.id", f"doc_{financial_scenario or 'default'}")
                    llm_span.set_attribute("retrieval.documents.0.document.score", 0.75)
                
                # Set status
                llm_span.set_status(Status(StatusCode.OK))
                
                # Set LLM latency (use calculated realistic value)
                llm_span.set_attribute("latency_ms", round(llm_latency_ms, 2))
                
                # Capture LLM span context for evaluation logging
                llm_span_context = llm_span.get_span_context()
                llm_span_id = format(llm_span_context.span_id, '016x')
            
            # Set AGENT latency (use calculated realistic value)
            agent_span.set_attribute("latency_ms", round(agent_latency_ms, 2))
            
            # Capture AGENT span context for evaluation logging
            agent_span_context = agent_span.get_span_context()
            agent_span_id = format(agent_span_context.span_id, '016x')
        
        # Set CHAIN output
        chain_span.set_attribute("output.value", response)
        chain_span.set_attribute("output.mime_type", "text/plain")
        chain_span.set_status(Status(StatusCode.OK))
        
        # Set CHAIN latency (use calculated realistic value)
        chain_span.set_attribute("latency_ms", round(chain_latency_ms, 2))
    
    return {
        "agent_name": agent_name,
        "is_poisoned": is_poisoned,
        "token_counts": {"prompt": prompt_tokens, "completion": completion_tokens, "total": total_tokens},
        "span_data": {
            "agent_span_id": agent_span_id,
            "llm_span_id": llm_span_id,
            "agent_span_kind": "AGENT",
            "llm_span_kind": "LLM",
            "scenario": financial_scenario or "stock_analysis",
            "query": query,
            "response": response
        }
    }
```

```python
# Session Evaluation Helper Function
# Processes CHAIN spans grouped by session to generate session evaluations

def generate_session_evaluations_from_chain_spans(span_df: pd.DataFrame) -> list:
    """
    Generate session evaluations from CHAIN spans.
    Groups by session_id, gets the oldest CHAIN span per session, and generates evaluations.
    
    Args:
        span_df: DataFrame containing span data with CHAIN spans
        
    Returns:
        list: List of evaluation dictionaries with session_eval.* columns
    """
    session_evaluations = []
    
    # Filter for CHAIN spans only
    chain_spans = span_df[span_df['openinference.span.kind'] == 'CHAIN'].copy()
    
    if len(chain_spans) == 0:
        return session_evaluations
    
    # Check if session_id column exists
    if 'session_id' not in chain_spans.columns:
        print("⚠️  No session_id column found in CHAIN spans. Skipping session evaluations.")
        return session_evaluations
    
    # Add timestamp if not present (use index as proxy for order)
    if 'timestamp' not in chain_spans.columns:
        chain_spans['timestamp'] = chain_spans.index
    
    # Remove rows with null session_id
    chain_spans = chain_spans[chain_spans['session_id'].notna()]
    
    if len(chain_spans) == 0:
        return session_evaluations
    
    # Group by session_id and get the oldest (first) CHAIN span per session
    # Sort by timestamp and take first per group
    chain_spans_sorted = chain_spans.sort_values('timestamp')
    oldest_chain_per_session = chain_spans_sorted.groupby('session_id').first().reset_index()
    
    print(f"  Processing {len(oldest_chain_per_session):,} unique sessions for session evaluations")
    
    # Generate session evaluations for oldest CHAIN spans
    for _, row in oldest_chain_per_session.iterrows():
        span_id = row.get('context.span_id', '')
        is_poisoned = row.get('is_poisoned', False)
        session_id = row.get('session_id', 'unknown')
        
        span_data = {
            'scenario': row.get('scenario', ''),
            'query': row.get('query', ''),
            'response': row.get('response', ''),
            'session_id': session_id
        }
        
        # Generate session evaluation
        # 80% chance of pass, 20% chance of fail
        # Poisoned data increases fail probability
        if is_poisoned:
            will_pass = random.random() < 0.35  # 35% pass rate for poisoned
        else:
            will_pass = random.random() < 0.82  # 82% pass rate for normal
        
        scenario = span_data.get('scenario', 'financial')
        
        if will_pass:
            label = "pass"
            score = 0
            explanations = [
                f"Session evaluation passed: The complete session for {scenario} scenario (session_id: {session_id}) successfully handled the user interaction. All components (agent routing, LLM responses, tool calls) worked together cohesively to deliver a satisfactory outcome.",
                f"Session validation successful: The end-to-end session flow for {scenario} was executed correctly. The chain orchestrated the agent and LLM components effectively, maintaining context and delivering appropriate responses throughout the interaction.",
                f"Session quality verified: The session (session_id: {session_id}) demonstrated proper coordination between all components for the {scenario} scenario. User query was processed correctly, and the final response met quality standards."
            ]
        else:
            label = "fail"
            score = 1
            explanations = [
                f"Session evaluation failed: The complete session for {scenario} scenario (session_id: {session_id}) encountered issues in coordination or execution. Components did not work together effectively, leading to suboptimal user experience or incorrect outcomes.",
                f"Session validation failed: The end-to-end session flow for {scenario} had coordination problems. The chain failed to properly orchestrate agent and LLM components, resulting in context loss, incorrect routing, or poor response quality.",
                f"Session quality check failed: The session (session_id: {session_id}) for {scenario} scenario showed misalignment between components. Agent decisions, LLM responses, or tool usage were inconsistent, leading to an unsatisfactory user experience."
            ]
        
        eval_data = {
            "label": label,
            "score": score,
            "explanation": random.choice(explanations)
        }
        
        session_evaluations.append({
            'context.span_id': span_id,
            'session_eval.label': eval_data['label'],
            'session_eval.score': eval_data['score'],
            'session_eval.explanation': eval_data['explanation']
        })
    
    return session_evaluations
```

```python
# Evaluation Generation Functions
# Generate synthetic evaluations for AGENT and LLM spans

import pandas as pd

def generate_agent_trajectory_accuracy_eval(span_data: dict, is_poisoned: bool) -> dict:
    """
    Generate AgentTrajectoryAccuracy evaluation for AGENT spans.
    
    Args:
        span_data: Dictionary containing span information
        is_poisoned: Whether the span has poisoned data (high token counts)
    
    Returns:
        dict: Evaluation data with label, score, and explanation
    """
    # 70% chance of pass, 30% chance of fail
    # Poisoned data increases fail probability
    if is_poisoned:
        will_pass = random.random() < 0.3  # 30% pass rate for poisoned
    else:
        will_pass = random.random() < 0.75  # 75% pass rate for normal
    
    if will_pass:
        label = "pass"
        score = 0  # 0 for pass
        explanations = [
            f"Agent trajectory accuracy check passed: The agent correctly identified and routed the {span_data.get('scenario', 'financial')} query to the appropriate handler. All decision points were logical and aligned with expected workflow.",
            f"Agent trajectory validation successful: The agent followed the correct sequence of operations for handling a {span_data.get('scenario', 'financial')} request. Tool selection and execution order were appropriate.",
            f"Agent behavior verified: The agent demonstrated proper trajectory by selecting relevant tools and maintaining context throughout the {span_data.get('scenario', 'financial')} interaction. No deviations from expected path detected."
        ]
    else:
        label = "fail"
        score = 1  # 1 for fail
        explanations = [
            f"Agent trajectory accuracy check failed: The agent deviated from expected workflow for {span_data.get('scenario', 'financial')} scenario. Incorrect tool selection or routing detected. Agent should have used different tools or followed alternative path.",
            f"Agent trajectory validation failed: The agent made suboptimal decisions during the {span_data.get('scenario', 'financial')} interaction. Tool execution order was incorrect, leading to inefficient processing.",
            f"Agent behavior misaligned: The agent failed to maintain proper context or made incorrect routing decisions for the {span_data.get('scenario', 'financial')} query. Trajectory does not match expected patterns for this scenario type."
        ]
    
    return {
        "label": label,
        "score": score,
        "explanation": random.choice(explanations)
    }

def generate_hallucination_eval(span_data: dict, is_poisoned: bool) -> dict:
    """
    Generate Hallucination evaluation for LLM spans.
    
    Args:
        span_data: Dictionary containing span information
        is_poisoned: Whether the span has poisoned data (high token counts)
    
    Returns:
        dict: Evaluation data with label, score, and explanation
    """
    # 80% chance of pass, 20% chance of fail
    # Poisoned data increases fail probability
    if is_poisoned:
        will_pass = random.random() < 0.4  # 40% pass rate for poisoned
    else:
        will_pass = random.random() < 0.85  # 85% pass rate for normal
    
    query = span_data.get('query', '')
    response = span_data.get('response', '')
    scenario = span_data.get('scenario', 'financial')
    
    if will_pass:
        label = "pass"
        score = 0  # 0 for pass
        explanations = [
            f"Hallucination check passed: The LLM response for the {scenario} query is factually consistent and does not contain fabricated information. All claims are supported by the provided context and align with expected financial data patterns.",
            f"No hallucination detected: The response accurately addresses the {scenario} query without introducing unsupported claims or fictional data. Content is grounded in realistic financial scenarios.",
            f"Response validation successful: The LLM output for the {scenario} interaction maintains factual accuracy. No evidence of hallucinated stock prices, account numbers, or transaction details. Response is coherent and contextually appropriate."
        ]
    else:
        label = "fail"
        score = 1  # 1 for fail
        explanations = [
            f"Hallucination detected: The LLM response contains fabricated or unsupported information for the {scenario} query. The response includes specific numbers, dates, or facts that cannot be verified against the provided context or standard financial data.",
            f"Factual inconsistency found: The response to the {scenario} query includes hallucinated details such as incorrect stock prices, invalid account numbers, or non-existent transaction IDs. The LLM appears to have generated plausible-sounding but false information.",
            f"Hallucination check failed: The LLM generated response for {scenario} contains unsubstantiated claims or fictional financial data. Specific metrics, valuations, or transaction details appear to be invented rather than derived from the provided context."
        ]
    
    return {
        "label": label,
        "score": score,
        "explanation": random.choice(explanations)
    }

def create_evaluations_from_span_data(span_df: pd.DataFrame) -> pd.DataFrame:
    """
    Create evaluation DataFrame from collected span data.
    
    Args:
        span_df: DataFrame containing span data with columns:
                 - context.span_id
                 - openinference.span.kind
                 - is_poisoned
                 - scenario
                 - query
                 - response
    
    Returns:
        pd.DataFrame: Evaluations DataFrame with appropriate schema
    """
    evaluations = []
    
    for _, row in span_df.iterrows():
        span_kind = row.get('openinference.span.kind', '')
        span_id = row.get('context.span_id', '')
        is_poisoned = row.get('is_poisoned', False)
        
        span_data = {
            'scenario': row.get('scenario', ''),
            'query': row.get('query', ''),
            'response': row.get('response', '')
        }
        
        if span_kind == 'AGENT':
            eval_data = generate_agent_trajectory_accuracy_eval(span_data, is_poisoned)
            evaluations.append({
                'context.span_id': span_id,
                'trace_eval.AgentTrajectoryAccuracy.label': eval_data['label'],
                'trace_eval.AgentTrajectoryAccuracy.score': eval_data['score'],
                'trace_eval.AgentTrajectoryAccuracy.explanation': eval_data['explanation']
            })
        elif span_kind == 'LLM':
            eval_data = generate_hallucination_eval(span_data, is_poisoned)
            evaluations.append({
                'context.span_id': span_id,
                'eval.Hallucination.label': eval_data['label'],
                'eval.Hallucination.score': eval_data['score'],
                'eval.Hallucination.explanation': eval_data['explanation']
            })
        # Drop all other span kinds (CHAIN, TOOL, etc.)
    
    return pd.DataFrame(evaluations)
```

#Full batch trace creation

```python
# Batch generation using the single trace function
# This cell generates multiple traces using the reusable create_single_trace() function

from opentelemetry import trace

tracer = trace.get_tracer(__name__)

# Number of traces to create - adjust as needed
NUM_SPANS = 1000

print(f"Generating {NUM_SPANS:,} traces (CHAIN -> AGENT -> LLM) with BatchSpanProcessor...")
print("Note: BatchSpanProcessor batches automatically. Smaller batches = faster exports.\n")

with timer():
    poisoned_count = 0
    
    # Collect span data in memory for evaluation generation
    span_data_list = []
    
    # Session management: Create small groups of traces with shared session IDs
    # Each session contains 2-5 traces
    SESSION_SIZE_MIN = 2
    SESSION_SIZE_MAX = 5
    current_session_id = None
    traces_in_current_session = 0
    session_size = 0
    
    # Create traces in batches to show progress and manage memory
    BATCH_SIZE = 10  # Process in batches of 1k for progress updates
    
    for batch_start in range(0, NUM_SPANS, BATCH_SIZE):
        batch_end = min(batch_start + BATCH_SIZE, NUM_SPANS)
        
        for i in range(batch_start, batch_end):
            # Start a new session if needed
            if current_session_id is None or traces_in_current_session >= session_size:
                # Generate new session ID
                current_session_id = f"session_{random.randint(100000, 999999)}"
                session_size = random.randint(SESSION_SIZE_MIN, SESSION_SIZE_MAX)
                traces_in_current_session = 0
            
            # Split traffic: 50% Financial Advisor, 50% Customer Service
            agent_type = random.choice(["financial_advisor", "customer_service"])
            
            if agent_type == "financial_advisor":
                # Financial advisor scenarios
                scenario = random.choice(["stock_analysis", "transaction_processing", "risk_assessment", "portfolio_analysis"])
                agent_name = "financial_advisor"
            else:
                # Customer service scenarios
                scenario = random.choice(["account_questions", "it_support"])
                agent_name = "customer_service"
            
            # Create trace using reusable function with session ID
            result = create_single_trace(
                tracer=tracer,
                agent_name=agent_name,
                financial_scenario=scenario,
                prompt_template_config=PROMPT_TEMPLATE_CONFIG,
                poisoned_config=POISONED_DATA_CONFIG,
                session_id=current_session_id
            )
            
            traces_in_current_session += 1
            
            # Collect span data for evaluation generation (only AGENT and LLM spans)
            if "span_data" in result:
                span_data = result["span_data"]
                is_poisoned = result["is_poisoned"]
                
                # Add AGENT span data
                span_data_list.append({
                    "context.span_id": span_data["agent_span_id"],
                    "openinference.span.kind": span_data["agent_span_kind"],
                    "is_poisoned": is_poisoned,
                    "scenario": span_data["scenario"],
                    "query": span_data["query"],
                    "response": span_data["response"]
                })
                
                # Add LLM span data
                span_data_list.append({
                    "context.span_id": span_data["llm_span_id"],
                    "openinference.span.kind": span_data["llm_span_kind"],
                    "is_poisoned": is_poisoned,
                    "scenario": span_data["scenario"],
                    "query": span_data["query"],
                    "response": span_data["response"]
                })
            
            if result["is_poisoned"]:
                poisoned_count += 1
        
        # Progress update
        if (batch_end % 1000 == 0) or (batch_end == NUM_SPANS):
            print(f"  Created {batch_end:,} / {NUM_SPANS:,} traces ({poisoned_count} poisoned)...")
    
    print(f"\nCompleted creating {NUM_SPANS:,} traces. Flushing remaining spans...")
    
    # Force flush to ensure all spans are exported
    # Calculate timeout based on number of spans
    base_timeout = 30000  # 30 seconds base
    timeout_per_10k_spans = 10000  # +10 seconds per 10k spans
    flush_timeout = base_timeout + (NUM_SPANS // 10000) * timeout_per_10k_spans
    flush_timeout = min(flush_timeout, 300000)  # Cap at 5 minutes max
    
    print(f"Flushing with timeout: {flush_timeout/1000:.0f} seconds")
    flush_success = trace.get_tracer_provider().force_flush(timeout_millis=flush_timeout)
    
    if flush_success:
        print(f"✅ Flush successful: All traces were exported within {flush_timeout/1000:.0f} seconds")
        print(f"   Total: {NUM_SPANS:,} traces, {poisoned_count} with poisoned data")
        print("   Note: This confirms traces were sent to Arize. Check the Arize dashboard to verify receipt.")
        print(f"\n   Collected {len(span_data_list):,} span records for evaluation generation")
        print("   Run the next cell to generate and log evaluations.")
    else:
        print(f"⚠️  Flush timeout: Some traces may not have been exported within {flush_timeout/1000:.0f} seconds")
        print("   Consider increasing the timeout or reducing batch size.")

# Store span_data_list outside timer context for use in evaluation cell
collected_span_data = span_data_list
```
